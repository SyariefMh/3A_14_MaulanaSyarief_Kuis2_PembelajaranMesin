```python
import numpy as np
import pandas as pd

df = pd.read_excel('data/dataset_quiz.xlsx')

df.head()
```

1. Lakukan pemilahan data training dan testing yang sesuai dan proporsional dari data tersebut! (10 poin)

```python
df.info()
```

```python
df.isnull().sum()
```

```python
modus_ketua = df["id_ketua"].mode()[0]
df["id_ketua"] = df["id_ketua"].replace(np.nan, modus_ketua)

modus_anggota1 = df["id_anggota1"].mode()[0]
df["id_anggota1"] = df["id_anggota1"].replace(np.nan, modus_anggota1)

modus_anggota2 = df["id_anggota2"].mode()[0]
df["id_anggota2"] = df["id_anggota2"].replace(np.nan, modus_anggota2)

modus_anggota3 = df["id_anggota3"].mode()[0]
df["id_anggota3"] = df["id_anggota3"].replace(np.nan, modus_anggota3)

mean_anggota4 = df["id_anggota4"].mean()
df["id_anggota4"] = df["id_anggota4"].replace(np.nan, mean_anggota4)

df.drop('id_anggota5', axis=1, inplace=True)

df
```

```python
df.isnull().sum()
```

2. Tentukan data validasi! (10 poin)

```python
from sklearn.model_selection import train_test_split

df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=3, stratify=df['keterlibatan_vltr'])

df_val, df_test = train_test_split(df_unseen, test_size=0.2, random_state=0, stratify=df_unseen['keterlibatan_vltr'])

print(f'Jumlah label data asli : \n{df.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data training : \n{df_train.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data validasi : \n{df_val.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data tes : \n{df_test.keterlibatan_vltr.value_counts()}')
```

3. Tentukan secara tepat kolom yang menjadi "label" (20 poin)

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

df['keterlibatan_vltr']= le.fit_transform(df['keterlibatan_vltr'])
df['skema']= le.fit_transform(df['skema'])
df['luaran_hki']= le.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki']= le.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki']= le.fit_transform(df['potensi_hki'])
df['luaran_lain']= le.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain']= le.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain']= le.fit_transform(df['status_luaran_lain'])
```

```python
x = df.iloc[:,6:16]
y = df['tahun']
x.head()
```

4. Terapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi  dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! (20 poin)

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)
print(y_pred_dt)

acc_dt = accuracy_score(y_test,y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

```python
from sklearn.svm import SVC

svm_poly = SVC(kernel='poly', degree=8)

svm_poly.fit(x_train, y_train)

y_pred_svm_poly = svm_poly.predict(x_test)
print(y_pred_svm_poly)

acc_svm_poly= accuracy_score(y_test, y_pred_svm_poly)

print("Test set accuracy : {:.2f}".format(acc_svm_poly))
print(f"Test st accuracy : {acc_svm_poly}")
```

5. Lakukan tunning hyperparameter berdasarkan hasil klasifikasi pada nomor 3 menggunakan metode Grid Search dan Random Search! Simpulkan parameter apa yang paling baik digunakan berdasarkan metode klasifikasi yang dipilih, dan berapa akurasinya. (40 poin)

```python
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt = DecisionTreeClassifier()

params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

grid = GridSearchCV(dt, param_grid=params, cv=cv)

%timeit grid.fit(x,y)

score = grid.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm = SVC()

params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

grid = GridSearchCV(svm, param_grid=params, cv=cv)

%timeit grid.fit(x,y)

score = grid.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dt = DecisionTreeClassifier()

params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

grid = RandomizedSearchCV(dt, params, cv=cv)

%timeit grid.fit(x,y)

score = grid.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

svm = SVC()

params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

grid = RandomizedSearchCV(svm, params, cv=cv)

%timeit grid.fit(x,y)

score = grid.score(x,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```
